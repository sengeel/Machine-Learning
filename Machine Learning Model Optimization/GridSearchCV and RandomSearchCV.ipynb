```python
# importing the dependencies
import ast
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
# number of rows and Columns in this dataset
data_frame.shape
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```


### 1 --> Benign

### 0 --> Malignant

### Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

# GridSearchCV

##### GridSearchCV is used for determining the best parameters for our model

```python
model = SVC()
```

```python
parameters = {
                'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
                'C' : [1, 5, 10, 20]
}
```

```python
type(parameters)
```

```python
classifier = GridSearchCV(model, param_grid=parameters, cv=5)
```

```python
classifier.fit(X, Y)
```

```python
df = pd.DataFrame(classifier.cv_results_)
```

```python
df
```

```python
# df['params'] = df['params'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

# Expand the 'params' column directly
params_df = pd.json_normalize(df['params'])

# Concatenate with the original DataFrame
df_expanded = pd.concat([df, params_df], axis=1).drop(columns=['params'])

df_expanded
```

```python
# best params

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# highest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
grid_search_results = df_expanded[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
grid_search_results
```

##### Highest Accuracy = 95.2%
##### Best Parameter = {'C':10, 'kernel':'linear'} 

## *RandomizedSearchCV*

```python
# loading the SVC model
model = SVC()
```

```python
parameters = {
                'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
                'C': [1, 5, 10, 20]
}
```

```python
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X, Y)
```

```python
df = pd.DataFrame(classifier.cv_results_)
```

```python
df
```

```python
# Expand the 'params' column directly
params_df = pd.json_normalize(df['params'])

# Concatenate with the original DataFrame
df_expanded = pd.concat([df, params_df], axis=1).drop(columns=['params'])

df_expanded
```

```python
# best params

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# higest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
randomized_search_result = df_expanded[['param_C','param_kernel','mean_test_score']]
```

```python
randomized_search_result
```

#### Highest Accuracy = 95.2%
#### Best Parameters = {'C':10, 'kernel':'linear'}
